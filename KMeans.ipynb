```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("/content/income.csv")
df.head()
```

```python
sns.scatterplot(x='Age', y='Income($)',data=df)
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=3)
km
```

```python
y_predicted=km.fit_predict(df[['Age','Income($)']])
```

```python
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
df0=df[df.cluster==0]
df1=df[df.cluster==1]
df2=df[df.cluster==2]
```

```python
sns.scatterplot(x='Age',y='Income($)',data=df0,color='green')
sns.scatterplot(x='Age',y='Income($)',data=df1,color='red')
sns.scatterplot(x='Age',y='Income($)',data=df2,color='black')
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df['Income($)']=scaler.fit_transform(df[['Income($)']])
```

```python
df.head()
```

```python
df['Age']=scaler.fit_transform(df[['Age']])
```

```python
df.head()
```

```python
df=df.drop(columns='cluster',axis=1)
```

```python
df.head()
```

```python
y_predicted=km.fit_predict(df[['Age','Income($)']])
```

```python
km=KMeans(n_clusters=3)
y_predicted=km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
df0=df[df.cluster==0]
df1=df[df.cluster==1]
df2=df[df.cluster==2]
```

```python
sns.scatterplot(x='Age',y='Income($)',data=df0,color='green')
sns.scatterplot(x='Age',y='Income($)',data=df1,color='red')
sns.scatterplot(x='Age',y='Income($)',data=df2,color='black')
sns.scatterplot(x=km.cluster_centers_[:,0], y=km.cluster_centers_[:,1], color='blue')
```

```python
km.cluster_centers_
```

```python
k_rng=range(1,10)
sse=[]
for k in k_rng:
  km=KMeans(n_clusters=k)
  km.fit(df[['Age','Income($)']])
  sse.append(km.inertia_)
```

```python
sse
```

```python
import matplotlib.pyplot as plt
plt.xlabel('k')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse) # 'k_rng' and 'sse' are lists, use plt.plot() directly.
```
