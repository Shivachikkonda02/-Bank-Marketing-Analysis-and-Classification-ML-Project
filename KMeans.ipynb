```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
data=load_iris()
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
from sklearn.cluster import KMeans
```

```python
wscc=[]

for i in range(2,21):
  km=KMeans(n_clusters=i,init='k-means++') ## To find out best cluster
  km.fit(df)
  wscc.append(km.inertia_)
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x=range(2,21),y=wscc,marker='o')
plt.xticks(range(2,21))
plt.show()
```

```python
kmn=KMeans(n_clusters=3)
df['predict']=kmn.fit_predict(df)
```

```python
df.head()
```

```python
sns.scatterplot(x='sepal length (cm)',y='sepal width (cm)',hue='predict',data=df)
```

```python
sns.pairplot(df,hue='predict')
```
